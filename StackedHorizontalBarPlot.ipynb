# Stacked Horizontal Bar Plot

```python
import numpy as np
import pandas as plt
import matplotlib.pyplot as plt
```

```python
# Sorted

high = np.array([ 337, 628, 722,  636,  649,  510,  351])
med  = np.array([ 969, 1950, 2077, 1886, 1905, 1479,  942])
low  = np.array([2866, 5643, 6425, 5521, 4897, 4845, 3179])
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

total = high + med + low
sorted_indices = np.argsort(total)

high_sorted = high[sorted_indices]
med_sorted = med[sorted_indices]
low_sorted = low[sorted_indices]
days_sorted = [days[i] for i in sorted_indices]

x = np.arange(1,8)

fig, ax = plt.subplots(figsize=(10,6))

ax.barh(x, high_sorted, label='High', color='#2270e6')
ax.barh(x, med_sorted, left=high_sorted, label='Medium', color='#a9b9d1')
ax.barh(x, low_sorted, left=high_sorted + med_sorted, label='Low', color='#ced2d9')

ax.set_yticklabels([''] + days_sorted)

ax.set_ylabel('Day of the week')
ax.set_title('Average number of listings per interest level', x=.235, y=1.1)

ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(axis='x', color='w')
ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)

ax.yaxis.set_label_coords(-0.105, 0.85)

ax.legend()

plt.show()
```
